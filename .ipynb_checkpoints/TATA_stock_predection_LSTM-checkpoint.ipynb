```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('NSE-TATAGLOBAL.csv')
df.head()
```

```python
training_set = df.iloc[:,1:2].values
```

```python
df.shape
```

```python
#feature Scalling 
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range=(0, 1))
training_set_scalled = sc.fit_transform(training_set)
```

```python
df.head()
```

```python
df.shape
```

```python
training_set_scalled.shape
```

```python
#Creating a data structure with 60 timesteps and 1 output 
X_train = []
y_train = []

for i in range(60, 2035):
    X_train.append(training_set_scalled[i-60:i, 0])
    y_train.append(training_set_scalled[i, 0])

X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
df.shape
```

```python
X_train.shape
```

```python
X_train.shape[0]
```

```python
#Reshaping 
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
X_train.shape
```

```python
#Part 2 - Building the RNN 

#Import the keras libraries and package 
from keras.models import Sequential 
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
input_shape = (X_train.shape[1], 1)
```

```python
input_shape
```

```python
#Installing the RNN 
regressor = Sequential() 
regressor.add(LSTM(units=50, return_sequences=True, input_shape = (X_train.shape[1], 1)))
regressor.add(Dropout(0.2))
```

```python
#Adding the Second LSTM layer and some dropout regularisation
regressor.add(LSTM(units=50, return_sequences=True))
regressor.add(Dropout(0.2))
```

```python
#Adding the Third LSTM layer and some dropout regularisation
regressor.add(LSTM(units=50, return_sequences=True))
regressor.add(Dropout(0.2))
```

```python
#Adding the fourth LSTM layer and some dropout regularisation
regressor.add(LSTM(units=50, return_sequences=True))
regressor.add(Dropout(0.2))
```

```python
#Adding the output Layers
regressor.add(Dense(units=1))

#Compiling the RNN 
regressor.compile(optimizer = 'rmsprop', loss = 'mean_squared_error' )
```

```python
#fitting the RNN to the training set 
regressor.fit(X_train, y_train, epochs = 10, batch_size=32)
```

```python
#part 3 - Making the prediction and visualizing the data 
#getting th real stock price of 2017
df_test= pd.read_csv('tatatest.csv')
real_stock_price = df_test.iloc[:, 1:2].values
```

```python
df_test.shape
```

```python
real_stock_price
```

```python
#Getting the predicted stock price of 2017 
df_total = pd.concat((df['Open'], df_test['Open']), axis=0)
```

```python
df_total
```

```python
inputs = df_total[len(df_total) - len(df_test) - 60 :].values
```

```python
len(inputs)
```

```python
inputs = inputs.reshape(-1, 1)
inputs = sc.transform(inputs)
X_test = [] 
for i in range (60, 308):
    X_test.append(inputs[i-60:i, 0])
```

```python
len(X_test)
```

```python
# Convert list to numpy array
X_test = np.array(X_test)

# Reshape to 3D for LSTM: (samples, timesteps, features)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

# Predict
predicted_stock_price = regressor.predict(X_test)

# Ensure 2D array for inverse_transform
predicted_stock_price = predicted_stock_price.reshape(-1, 1)

# Inverse transform to original scale
predicted_stock_price = sc.inverse_transform(predicted_stock_price)
```
