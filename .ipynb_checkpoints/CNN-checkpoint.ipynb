```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import random
from skimage import exposure
from skimage.util import random_noise
from skimage import transform
from cv2 import resize
```

```python
img = mpimg.imread('cat1.jpg')
plt.imshow(img)
```

```python
img_rescale = resize(img, (1500,1500))
plt.imshow(img_rescale)
```

```python
img_rescale = resize(img, (50,50))
plt.imshow(img_rescale)
```

```python
horizontal_flip = np.fliplr(img)
plt.imshow(horizontal_flip)
```

```python
vertical_flip = np.flipud(img)
plt.imshow(vertical_flip)
```

```python
trans_img = transform.rotate(img, random.uniform(40, 40))
plt.imshow(trans_img)
```

```python
#Adding Noise 
img_noise = random_noise(img, mode='s&p', clip=True)
plt.imshow(img_noise)
```

```python
#Adding Noise 
img_noise = random_noise(img, mode='gaussian', clip=True)
plt.imshow(img_noise)
```

```python
#Adding Noise 
img_noise = random_noise(img, mode='poisson', clip=True)
plt.imshow(img_noise)
```
