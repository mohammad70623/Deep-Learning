```python
#import the keras libraries and p__package 
import keras
from keras.models import Sequential
from keras.layers import Dense , Input
```

```python
#Initialiging the ANN 
classifier = Sequential()

#Units: 
#initializer: how your weight are updated
#relu activation function
#input_dim: imput shape
#adding the input layer and the first hidden layer 
classifier.add(Input(shape =(10,)))
classifier.add(Dense(units=6, kernel_initializer='uniform', activation='relu'))
#Adding the second hidden layer 
classifier.add(Dense(units=6, kernel_initializer='uniform', activation='relu'))
#Adding the third hidden layer 

classifier.add(Dense(units=6, kernel_initializer='uniform', activation='relu'))

#Adding the output Layer 
classifier.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))
```

```python
len(x_train)
```

```python
#Compiling the ANN 
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
y_train[0]
```

```python
#Fitting the ANN to the training set 
classifier.fit(x_train, y_train, batch_size = 10, epochs = 20)
```

```python
#making prediction and evaluate the model 
#Predicting the test set result 
y_predict_ann = classifier.predict(x_test)
y_predict_ann =(y_predict_ann > 0.5)
cm3 = confusion_matrix(y_test, y_predict_ann)
```

Parameter optimization

```python
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score
from keras.models import Sequential
from keras.layers import Dense, Input
```

```python
def build_classifier():
    classifier = Sequential()
    classifier.add(Input(shape=(10,)))
    classifier.add(Dense(units=6, kernel_initializer='uniform', activation='relu'))
    classifier.add(Dense(units=6, kernel_initializer='uniform', activation='relu'))
    classifier.add(Dense(units=6, kernel_initializer='uniform', activation='relu'))
    classifier.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))
    classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return classifier

# Wrap Keras model for scikit-learn
classifier = KerasClassifier(build_fn=build_classifier, batch_size=10, epochs=10)

# K-Fold Cross Validation
accuracies = cross_val_score(estimator=classifier, X=x_train, y=y_train, cv=10, n_jobs=-1)

mean = accuracies.mean()
variance = accuracies.std()

```

```python
print("Mean Accuracy:", mean)
print("Standard Deviation:", variance)
```
