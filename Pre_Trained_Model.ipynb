```python
from keras.applications import VGG16
from tensorflow.keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input, decode_predictions
import numpy as np
```

```python
#oad the vgg16 pre trained model and specify the input shape 
model = VGG16(weights='imagenet', include_top=True, input_shape=(224, 224,3))
```

```python
#Laoding an image and preprocess it for the model 
img_path = 'microphone.jpg'
img = image.load_img(img_path, target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis = 0)
x = preprocess_input(x)
```

```python
#Use the model to predic the class 
pred = model.predict(x)
```

```python
pred
```

```python
#Get the top 5 prediction wwith class name 
decode_predict = decode_predictions(pred, top = 10)[0]
```

```python
for pred in decode_predict:
    print(pred[1], ":" ,pred[2])
```
